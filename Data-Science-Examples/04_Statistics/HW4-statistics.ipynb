## Random sampling presentation
Problem 1. Choosing a random sampling of 10 numbers and counting them. Find frequency of each number.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats
```

```python
items = [8, 3, 1, 8, 6, 5, 10, 1, 8, 1]

def count_intems(items):
    counts = {}
    for item in items:
        if item not in counts:
            counts[item] = 1
        else:
            counts[item] += 1
            
    length = len(items)
    for item, count in counts.items():
        print(item, "frequency/variance:", item/length)
        print(count)
    return counts

print(count_intems(items))
```

```python
Draw a histogram
```

```python
plt.figure()
plt.hist(items)
plt.xlabel('Numbers')
plt.ylabel('Counts')
plt.show()

plt.figure()
plt.hist(items, normed = 1)  #frequencies or probabilities
plt.xlabel('Numbers')
plt.ylabel('Frequencies')
plt.show()

plt.figure()
plt.hist(items, normed = 1, cumulative = True)  #cumulative function of plutnostta
plt.xlabel('Numbers')
plt.ylabel('Probability')
plt.show()
```

Display with bins

```python
# 20 random numbers in [1; 500), by default with 10 intervals
x = np.random.uniform(1, 500, 20)  # chislata sa ravnomerno razpredeleni
plt.hist(x)
plt.show()

plt.hist(x, bins = 5)
plt.show()

plt.hist(x, bins = 100)  #
plt.show()

plt.figure()
plt.hist(x, bins = 100, cumulative = True)  #cumulative
plt.show()
```
